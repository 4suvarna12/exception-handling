9.how can you read a file line by line and store its content in a list in Python

```python
import logging

def configure_logger(log_file="file_errors.log", level=logging.ERROR):
    """Configures the logging system to write errors to a file.

    Args:
        log_file (str): The name of the log file.
        level (int): The logging level.
    """
    logging.basicConfig(
        filename=log_file,
        level=level,
        format='%(asctime)s - %(levelname)s - %(message)s'
    )

def read_file_lines(filename):
    """
    Reads a file line by line and stores the content in a list.
    Handles potential FileNotFoundError and other exceptions.

    Args:
        filename (str): The name of the file to read.

    Returns:
        list: A list where each element is a line from the file, or None on error.
              The lines include the newline character ('\n').
    """
    lines = []
    try:
        with open(filename, 'r') as file:
            for line in file:
                lines.append(line)
        logging.info(f"Successfully read file: {filename}")
        return lines
    except FileNotFoundError:
        logging.error(f"File not found: {filename}")
        print(f"Error: File not found - {filename}")
        return None
    except Exception as e:
        logging.error(f"An error occurred while reading {filename}: {e}")
        print(f"An unexpected error occurred: {e}")
        return None

def main():
    """
    Main function to demonstrate reading a file line by line.
    """
    configure_logger()  # Initialize logging

    # Test case:
    filename = "my_file.txt"
    # Create a sample file (if it doesn't exist)
    try:
        with open(filename, 'w') as f:
            f.write("Line 1\nLine 2\nLine 3\n")
    except Exception as e:
        print(f"Error creating sample file: {e}")
        logging.error(f"Error creating sample file: {e}")

    lines = read_file_lines(filename)
    if lines:
        print(f"Lines from {filename}:")
        for line in lines:
            print(f"'{line.strip()}'")  # Use strip() to remove leading/trailing spaces and newlines

if __name__ == "__main__":
    main()
```

10.How can you append data to an existing file in Python

```python
import logging

def configure_logger(log_file="file_errors.log", level=logging.ERROR):
    """Configures the logging system to write errors to a file.

    Args:
        log_file (str): The name of the log file.
        level (int): The logging level.
    """
    logging.basicConfig(
        filename=log_file,
        level=level,
        format='%(asctime)s - %(levelname)s - %(message)s'
    )

def append_to_file(filename, data):
    """
    Appends data to an existing file.  Handles potential errors during the file operation.

    Args:
        filename (str): The name of the file to append to.
        data (str): The data to append to the file.
    """
    try:
        with open(filename, 'a') as file:
            file.write(data)
        logging.info(f"Successfully appended {len(data)} bytes to {filename}")
    except Exception as e:
        logging.error(f"An error occurred while appending to {filename}: {e}")
        print(f"Error: An error occurred while appending to {filename}: {e}")

def main():
    """
    Main function to demonstrate appending data to a file.
    """
    configure_logger()

    filename = "my_file.txt"  # Use the same filename as in the previous example
    # Ensure the file exists before appending
    try:
        with open(filename, 'w') as f:
            f.write("Initial content\n")
    except Exception as e:
        print(f"Error creating file: {e}")
        logging.error(f"Error creating file: {e}")
        return

    append_to_file(filename, "Appended line 1\n")
    append_to_file(filename, "Appended line 2\n")

    print(f"Data appended to {filename}.  Check the file contents.")

    # Read and print the file contents to demonstrate the append
    try:
        with open(filename, 'r') as file:
            contents = file.read()
        print(f"\nContents of {filename} after appending:")
        print(contents)
    except Exception as e:
        print(f"Error reading file: {e}")
        logging.error(f"Error reading file: {e}")

if __name__ == "__main__":
    main()
```

```python
import logging
import os.path  # Import the os.path module

def configure_logger(log_file="app_exceptions.log", level=logging.ERROR):
    """Configures the logging system to write errors to a file.

    Args:
        log_file (str): The name of the log file.
        level (int): The logging level.
    """
    logging.basicConfig(
        filename=log_file,
        level=level,
        format='%(asctime)s - %(levelname)s - %(message)s'
    )

def read_file(filename):
    """
    Reads and prints the contents of a file, first checking if the file exists.
    Handles potential FileNotFoundError and other exceptions during file operations.

    Args:
        filename (str): The name of the file to read.
    """
    if not os.path.exists(filename):
        logging.error(f"File not found: {filename}")
        print(f"Error: File not found - {filename}")
        return  # Exit the function if the file doesn't exist

    try:
        with open(filename, 'r') as file:
            contents = file.read()
            print(f"Contents of {filename}:")
            print(contents)
        logging.info(f"Successfully read file: {filename}")
    except Exception as e:
        logging.error(f"An error occurred while reading {filename}: {e}")
        print(f"An unexpected error occurred: {e}")

def main():
    """
    Main function to demonstrate file reading with the existence check.
    """
    configure_logger()

    # Test cases:
    read_file("existing_file.txt")  # Should work if the file exists
    read_file("nonexistent_file.txt")  # Will be handled by the os.path.exists() check

    print("Program finished.")

if __name__ == "__main__":
    main()
```

17.Write a Python program to create and write a list of numbers to a file, one number per line

```python
import logging
import os

def configure_logger(log_file="file_operations.log", level=logging.INFO):
    """Configures the logging system to write to a file.

    Args:
        log_file (str): The name of the log file.
        level (int): The logging level.
    """
    # Create the log file directory if it doesn't exist
    log_dir = os.path.dirname(log_file)
    if log_dir and not os.path.exists(log_dir):
        os.makedirs(log_dir)

    logging.basicConfig(
        filename=log_file,
        level=level,
        format='%(asctime)s - %(levelname)s - %(message)s'
    )
    console_handler = logging.StreamHandler()
    console_handler.setLevel(level)
    formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
    console_handler.setFormatter(formatter)
    logging.getLogger('').addHandler(console_handler)

def write_list_to_file(numbers, filename="numbers.txt"):
    """
    Writes a list of numbers to a file, one number per line.

    Args:
        numbers (list): A list of numbers.
        filename (str): The name of the file to write to.  Defaults to "numbers.txt".
    """
    if not isinstance(numbers, list):
        logging.error("Input 'numbers' is not a list.")
        print("Error: Input 'numbers' must be a list.")
        return

    if not numbers:  # Check for empty list
        logging.warning("Input list 'numbers' is empty.  No data written to file.")
        print("Warning: Input list is empty. No data written to file.")
        return

    try:
        with open(filename, 'w') as file:
            for number in numbers:
                file.write(str(number) + "\n")
        logging.info(f"Successfully wrote {len(numbers)} numbers to file: {filename}")
        print(f"Successfully wrote {len(numbers)} numbers to file: {filename}")
    except Exception as e:
        logging.error(f"An error occurred while writing to {filename}: {e}")
        print(f"Error: An error occurred while writing to the file: {e}")

def main():
    """
    Main function to demonstrate writing a list of numbers to a file.
    """
    configure_logger()

    # Create a sample list of numbers
    numbers = [1, 2, 3, 4, 5, 10, 20, 30, 40, 50]

    # Write the list to a file
    write_list_to_file(numbers)

    # Example with a different filename
    write_list_to_file(numbers, "my_numbers.txt")

    # Example with an empty list
    write_list_to_file([])

    # Example with invalid input
    write_list_to_file("not a list")

    logging.info("Program finished.")

if __name__ == "__main__":
    main()
```

18.How would you implement a basic logging setup that logs to a file with rotation after 1MB

```python
import logging
import logging.handlers
import os

def configure_rotating_logger(log_file="rotated_app.log", level=logging.INFO, max_bytes=1024*1024, backup_count=5):
    """Configures a rotating file logger.

    Args:
        log_file (str): The name of the log file.
        level (int): The logging level.
        max_bytes (int): The maximum size of the log file before it rotates (default: 1MB).
        backup_count (int): The number of backup files to keep (default: 5).
    """
    # Create the log file directory if it doesn't exist
    log_dir = os.path.dirname(log_file)
    if log_dir and not os.path.exists(log_dir):
        os.makedirs(log_dir)

    # Create a rotating file handler
    rotating_handler = logging.handlers.RotatingFileHandler(
        log_file,
        maxBytes=max_bytes,
        backupCount=backup_count
    )

    # Set the logging format
    formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
    rotating_handler.setFormatter(formatter)

    # Get the root logger and add the handler
    logger = logging.getLogger('')  # Get the root logger
    logger.setLevel(level)
    logger.addHandler(rotating_handler)

    # Also log to console
    console_handler = logging.StreamHandler()
    console_handler.setLevel(level)
    console_handler.setFormatter(formatter)
    logger.addHandler(console_handler)

def main():
    """Main function to demonstrate logging with rotation."""
    configure_rotating_logger()  # Initialize the rotating logger

    # Log some messages to demonstrate rotation
    for i in range(20):
        logging.info(f"Message {i}: This is a log message.  This message is repeated to generate enough text to cause a rotation. The current counter is {i}")
        logging.warning(f"Warning message {i}")
        logging.error(f"Error message {i}")

    logging.info("Program finished.")

if __name__ == "__main__":
    main()
```

19.Write a program that handles both IndexError and KeyError using a try-except block

```python
import logging
import os

def configure_logger(log_file="app_exceptions.log", level=logging.ERROR):
    """Configures the logging system to write errors to a file.

    Args:
        log_file (str): The name of the log file.
        level (int): The logging level.
    """
    # Create the log file directory if it doesn't exist
    log_dir = os.path.dirname(log_file)
    if log_dir and not os.path.exists(log_dir):
        os.makedirs(log_dir)

    logging.basicConfig(
        filename=log_file,
        level=level,
        format='%(asctime)s - %(levelname)s - %(message)s'
    )
    console_handler = logging.StreamHandler()
    console_handler.setLevel(level)
    formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
    console_handler.setFormatter(formatter)
    logging.getLogger('').addHandler(console_handler)

def access_data(data, index=None, key=None):
    """
    Accesses data from a list or dictionary using either an index or a key.
    Handles IndexError and KeyError exceptions.

    Args:
        data (list or dict): The list or dictionary to access.
        index (int, optional): The index to access if data is a list. Defaults to None.
        key (str, optional): The key to access if data is a dict. Defaults to None.
    """
    try:
        if isinstance(data, list) and index is not None:
            value = data[index]
            print(f"Accessed list at index {index}: {value}")
            logging.info(f"Accessed list at index {index}: {value}")
            return value
        elif isinstance(data, dict) and key is not None:
            value = data[key]
            print(f"Accessed dictionary with key '{key}': {value}")
            logging.info(f"Accessed dictionary with key '{key}': {value}")
            return value
        else:
            message = "Error: Either 'index' or 'key' must be provided, and 'data' must be a list or dict."
            print(message)
            logging.error(message)
            return None
    except IndexError as e:
        print(f"Error: Index {index} is out of bounds for list. Error Details: {e}")
        logging.error(f"IndexError: Index {index} is out of bounds. Error Details: {e}")
        return None
    except KeyError as e:
        print(f"Error: Key '{key}' not found in dictionary. Error Details: {e}")
        logging.error(f"KeyError: Key '{key}' not found. Error Details: {e}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        logging.error(f"An unexpected error occurred: {e}")
        return None

def main():
    """
    Main function to demonstrate IndexError and KeyError handling.
    """
    configure_logger()

    # Test cases with a list
    my_list = [10, 20, 30, 40, 50]
    access_data(my_list, index=2)  # Accessing a valid index
    access_data(my_list, index=5)  # Accessing an invalid index (IndexError)
    access_data(my_list, key="invalid") #Should raise an error

    # Test cases with a dictionary
    my_dict = {"a": 1, "b": 2, "c": 3}
    access_data(my_dict, key="b")  # Accessing a valid key
    access_data(my_dict, key="d")  # Accessing an invalid key (KeyError)
    access_data(my_dict, index=0) #Should raise an error

    # Test cases with invalid input
    access_data(my_list)  # Missing both index and key
    access_data("not a list or dict", index=0)  # Invalid data type

    logging.info("Program finished.")

if __name__ == "__main__":
    main()
```

20.How would you open a file and read its contents using a context manager in Python

```python
import logging
import os

def configure_logger(log_file="file_operations.log", level=logging.INFO):
    """Configures the logging system to write to a file.

    Args:
        log_file (str): The name of the log file.
        level (int): The logging level.
    """
    # Create the log file directory if it doesn't exist
    log_dir = os.path.dirname(log_file)
    if log_dir and not os.path.exists(log_dir):
        os.makedirs(log_dir)

    logging.basicConfig(
        filename=log_file,
        level=level,
        format='%(asctime)s - %(levelname)s - %(message)s'
    )
    console_handler = logging.StreamHandler()
    console_handler.setLevel(level)
    formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
    console_handler.setFormatter(formatter)
    logging.getLogger('').addHandler(console_handler)


def read_file_with_context_manager(filename):
    """Reads the content of a file using a context manager (with statement).

    Args:
        filename (str): The name of the file to read.
    """
    try:
        with open(filename, 'r') as file:
            contents = file.read()
        logging.info(f"Successfully read file: {filename}")
        print(f"Contents of {filename}:\n{contents}")
        return contents
    except FileNotFoundError:
        logging.error(f"File not found: {filename}")
        print(f"Error: File not found - {filename}")
        return None
    except Exception as e:
        logging.error(f"An error occurred while reading {filename}: {e}")
        print(f"An unexpected error occurred: {e}")
        return None



def main():
    """
    Main function to demonstrate reading a file using a context manager.
    """
    configure_logger()

    # Create a sample file for demonstration
    sample_file = "sample.txt"
    try:
        with open(sample_file, "w") as f:
            f.write("This is a sample file.\nIt contains multiple lines.\n")
    except Exception as e:
        print(f"Error creating sample file: {e}")
        logging.error(f"Error creating sample file: {e}")
        return

    # Read the file using the context manager
    read_file_with_context_manager(sample_file)

    # Try to read a non-existent file
    read_file_with_context_manager("nonexistent.txt")

    logging.info("Program finished.")



if __name__ == "__main__":
    main()
```

21.Write a Python program that reads a file and prints the number of occurrences of a specific word

```python
import logging
import os
import re  # Import the regular expression module

def configure_logger(log_file="file_operations.log", level=logging.INFO):
    """Configures the logging system to write to a file.

    Args:
        log_file (str): The name of the log file.
        level (int): The logging level.
    """
    # Create the log file directory if it doesn't exist
    log_dir = os.path.dirname(log_file)
    if log_dir and not os.path.exists(log_dir):
        os.makedirs(log_dir)

    logging.basicConfig(
        filename=log_file,
        level=level,
        format='%(asctime)s - %(levelname)s - %(message)s'
    )
    console_handler = logging.StreamHandler()
    console_handler.setLevel(level)
    formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
    console_handler.setFormatter(formatter)
    logging.getLogger('').addHandler(console_handler)

def count_word_occurrences(filename, word):
    """Reads a file and counts the number of occurrences of a specific word.

    Args:
        filename (str): The name of the file to read.
        word (str): The word to count.

    Returns:
        int: The number of times the word appears in the file, or 0 if the file
             does not exist or an error occurs.
    """
    try:
        with open(filename, 'r') as file:
            text = file.read()
            # Use regular expressions for case-insensitive counting
            pattern = rf"\b{re.escape(word)}\b"  # \b for word boundaries, re.escape for special chars
            count = len(re.findall(pattern, text, re.IGNORECASE))
        logging.info(f"Successfully read file: {filename}")
        print(f"The word '{word}' appears {count} times in the file '{filename}'.")
        return count
    except FileNotFoundError:
        logging.error(f"File not found: {filename}")
        print(f"Error: File not found - {filename}")
        return 0
    except Exception as e:
        logging.error(f"An error occurred while reading {filename}: {e}")
        print(f"An unexpected error occurred: {e}")
        return 0

def main():
    """
    Main function to demonstrate counting word occurrences in a file.
    """
    configure_logger()

    # Create a sample file for demonstration
    sample_file = "sample.txt"
    try:
        with open(sample_file, "w") as f:
            f.write("This is a sample file.  The word 'sample' appears multiple times, Sample, sample.\n"
                    "It contains multiple lines.  We will count the occurrences of the word 'sample'.")
    except Exception as e:
        print(f"Error creating sample file: {e}")
        logging.error(f"Error creating sample file: {e}")
        return

    # Count occurrences of the word "sample"
    count = count_word_occurrences(sample_file, "sample")
    logging.info(f"Counted 'sample' in {sample_file}: {count}")

    # Count occurrences of a word not in the file
    count = count_word_occurrences(sample_file, "nonexistent")
    logging.info(f"Counted 'nonexistent' in {sample_file}: {count}")

    # Try to read a non-existent file
    count = count_word_occurrences("nonexistent.txt", "word")
    logging.info(f"Counted 'word' in nonexistent.txt: {count}") # Will be 0

    logging.info("Program finished.")


if __name__ == "__main__":
    main()
```

1.What is the difference between interpreted and compiled languages
- Compiled Languages
Translation: The source code is translated into machine code (or an intermediate language) before the program is executed. This translation is done by a tool called a compiler.

- Execution: The compiled code (the machine code) is then executed directly by the computer's CPU.

- Process: Source Code -> Compiler -> Machine Code -> Execution

- Examples: C, C++, Java, Rust

- Interpreted Languages
Translation: The source code is translated into machine code line by line during the execution of the program. This translation is done by a tool called an interpreter.

- Execution: The interpreter reads a line of code, translates it, and executes it immediately. This process is repeated for each line of code.

- Process: Source Code -> Interpreter -> Execution (line by line)

- Examples: Python, JavaScript, Ruby

12.What are the basic steps involved in exception handling in Python?

The code that might raise an exception is placed inside a try block.

This block tells Python to "try" running this code, and if an error occurs, go to the except block.

- Except Block(s):

If an exception occurs within the try block, the program's control is transferred to the except block.

The code within the except block is executed to handle the exception.

You can have multiple except blocks to handle different types of exceptions, allowing you to respond to specific errors in different ways.

- Optional Else Block:

An optional else block can be included.

The code within the else block is executed if no exception occurs in the try block.

This is useful for code that should only run if the try block succeeds.

- Optional Finally Block:

An optional finally block can be included.

The code within the finally block is always executed, regardless of whether an exception occurred or not.

This is often used for cleanup operations, such as closing files or releasing resources, that should be performed no matter what happens
